```python
rows = []
with open('_data\Customers.txt', encoding = 'utf-8') as f:
    line = f.readline()
    while line:
        #print(line)
        columns = line.split('\t')
        for i, column in enumerate(columns):
            columns[i] = column.strip()
        rows.append(columns)
        line = f.readline()
```

```python
rows
```

```python
for i, row in enumerate(rows):
    # prvni radek je nazev promenne
    if i > 0:
        #print(row[0], row[1], row[2], row[3], row[4], row[5], row[6])
        print('''INSERT INTO sqlw3s.Customers ({0}, {1}, {2}, {3}, {4}, {5}, {6}) VALUES ({7}, "{8}", "{9}", "{10}", "{11}", "{12}", "{13}");'''.format(rows[0][0], rows[0][1], rows[0][2], rows[0][3], rows[0][4], rows[0][5], rows[0][6], \
        rows[i][0], rows[i][1], rows[i][2], rows[i][3], rows[i][4], rows[i][5], rows[i][6]))
        cur.execute('''INSERT INTO sqlw3s.Customers ({0}, {1}, {2}, {3}, {4}, {5}, {6}) VALUES ({7}, "{8}", "{9}", "{10}", "{11}", "{12}", "{13}");'''.format(rows[0][0], rows[0][1], rows[0][2], rows[0][3], rows[0][4], rows[0][5], rows[0][6], \
        rows[i][0], rows[i][1], rows[i][2], rows[i][3], rows[i][4], rows[i][5], rows[i][6]))
```
