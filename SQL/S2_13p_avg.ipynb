```python
from MySQLdb import connect
import pandas as pd
```

```python
# Connect to your MySQLdb
conn = connect(host='localhost', user='root')
#conn = connect(host='127.0.0.1', user='root', password = 'root', port=8890)
```

```python
# Open a cursor to perform database operations
cur = conn.cursor()
```

### The SQL AVG() Function
- The AVG() function returns the average value of a numeric column.

```python
# Find the average price of all products:
cur.execute('''SELECT AVG(Price) FROM sqlw3s.Products;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Note: NULL values are ignored.

>Syntax:

>SELECT AVG(column_name)
>
>FROM table_name
>
>WHERE condition;

Add a Where Clause:
- You can add a WHERE clause to specify conditions:

```python
# Return the average price of products in category 1:
cur.execute('''SELECT AVG(Price) FROM sqlw3s.Products WHERE CategoryID = 1;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Use an Alias:
- Give the AVG column a name by using the AS keyword.

```python
# Name the column "average price":
cur.execute('''SELECT AVG(Price) AS "average price" FROM sqlw3s.Products;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Higher Than Average:
- To list all records with a higher price than average, we can use the AVG() function in a sub query:

```python
# Return all products with a higher price than the average price:
cur.execute('''SELECT * FROM sqlw3s.Products
WHERE price > (SELECT AVG(price) FROM sqlw3s.Products);''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```
