```python
from MySQLdb import connect
import pandas as pd
```

```python
# Connect to your MySQLdb
conn = connect(host='localhost', user='root')
```

```python
# Open a cursor to perform database operations
cur = conn.cursor()
```

The SQL ORDER BY
- The ORDER BY keyword is used to sort the result-set in ascending or descending order.

```python
# Sort the products by price:
cur.execute('''SELECT * FROM sqlw3s.Products ORDER BY Price;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

>Syntax
>
>SELECT column1, column2, ...
>FROM table_name
>ORDER BY column1, column2, ... ASC|DESC;

DESC
- The ORDER BY keyword sorts the records in ascending order by default. To sort the records in descending order, use the DESC keyword.

```python
# Sort the products from highest to lowest price:
cur.execute('''SELECT * FROM sqlw3s.Products ORDER BY Price DESC;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Order Alphabetically
- For string values the ORDER BY keyword will order alphabetically:

```python
# Sort the products alphatbetically by ProductName:
cur.execute('''SELECT * FROM sqlw3s.Products ORDER BY ProductName;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Alphabetically DESC
- To sort the table reverse alphabetically, use the DESC keyword:

```python
# Sort the products by ProductName in reverse order:
cur.execute('''SELECT * FROM sqlw3s.Products ORDER BY ProductName DESC;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

ORDER BY Several Columns
- The following SQL statement selects all customers from the "Customers" table, sorted by the "Country" and the "CustomerName" column. This means that it orders by Country, but if some rows have the same Country, it orders them by CustomerName:

```python
# Example:
cur.execute('''SELECT * FROM sqlw3s.Customers ORDER BY Country, CustomerName;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Using Both ASC and DESC
- The following SQL statement selects all customers from the "Customers" table, sorted ascending by the "Country" and descending by the "CustomerName" column:

```python
# Example:
cur.execute('''SELECT * FROM sqlw3s.Customers ORDER BY Country ASC, CustomerName DESC;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```
