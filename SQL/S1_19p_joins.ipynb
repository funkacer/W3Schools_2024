```python
# Open a cursor to perform database operations
cur = conn.cursor()
```

### SQL JOIN
- A JOIN clause is used to combine rows from two or more tables, based on a related column between them.f that query.
- An alias is created with the AS keyword.

```python
# Let's look at a selection from the "Orders" table:
cur.execute('''SELECT * FROM sqlw3s.Orders;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

```python
# Then, look at a selection from the "Customers" table:
cur.execute('''SELECT * FROM sqlw3s.Customers;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Notice that the "CustomerID" column in the "Orders" table refers to the "CustomerID" in the "Customers" table.
- The relationship between the two tables above is the "CustomerID" column.

```python
# Then, we can create the following SQL statement (that contains an INNER JOIN), that selects records that have matching values in both tables:
cur.execute('''SELECT Orders.OrderID, Customers.CustomerName, Orders.OrderDate
FROM sqlw3s.Orders
INNER JOIN sqlw3s.Customers ON Orders.CustomerID=Customers.CustomerID;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

#### Different Types of SQL JOINs
Here are the different types of the JOINs in SQL:

- ${\color{red}(INNER) JOIN}$: Returns records that have matching values in both tables
- ${\color{red}LEFT (OUTER) JOIN}$: Returns all records from the left table, and the matched records from the right table
- ${\color{red}RIGHT (OUTER) JOIN}$: Returns all records from the right table, and the matched records from the left table
- ${\color{red}FULL (OUTER) JOIN}$: Returns all records when there is a match in either left or right table
