```python
from MySQLdb import connect
import pandas as pd
```

```python
# Connect to your MySQLdb
conn = connect(host='localhost', user='root')
```

```python
# Open a cursor to perform database operations
cur = conn.cursor()
```

The NOT Operator
- The NOT operator is used in combination with other operators to give the opposite result, also called the negative result.
- In the select statement below we want to return all customers that are NOT from Spain:

```python
# Select only the customers that are NOT from Spain:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE NOT Country = 'Spain';''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

In the example above, the NOT operator is used in combination with the = operator, but it can be used in combination with other comparison and/or logical operators. See examples below.

> Syntax
>
>SELECT column1, column2, ...
>FROM table_name
>WHERE NOT condition;

NOT LIKE

```python
# Select customers that does not start with the letter 'A':
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE CustomerName NOT LIKE 'A%';''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

NOT BETWEEN

```python
# Select customers with a customerID not between 10 and 60:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE CustomerID NOT BETWEEN 10 AND 60;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

NOT IN

```python
# Select customers that are not from Paris or London:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE City NOT IN ('Paris', 'London');''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

NOT Greater Than

```python
# Select customers with a CustomerId not greater than 50:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE NOT CustomerID > 50;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Note: There is a not-greater-then operator: !> that would give you the same result.

NOT Less Than

```python
# Select customers with a CustomerID not less than 50:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE NOT CustomerId < 50;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Note: There is a not-less-then operator: !< that would give you the same result.
