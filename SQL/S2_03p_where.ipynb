The SQL WHERE Clause
- The WHERE clause is used to filter records.
- It is used to extract only those records that fulfill a specified condition.

```python
# Select all customers from Mexico:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE Country='Mexico';''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

>Syntax
>
>SELECT column1, column2, ...
>FROM table_name
>WHERE condition;

Note: The WHERE clause is not only used in SELECT statements, it is also used in UPDATE, DELETE, etc.!

Text Fields vs. Numeric Fields
- SQL requires single quotes around text values (most database systems will also allow double quotes).
- However, numeric fields should not be enclosed in quotes:

```python
# Example:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE CustomerID=1;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

Operators in The WHERE Clause
- You can use other operators than the = operator to filter the search.

```python
# Select all customers with a CustomerID greater than 80:
cur.execute('''SELECT * FROM sqlw3s.Customers WHERE CustomerID > 80;''')
df = pd.DataFrame(tuple(t) for t in cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

The following operators can be used in the WHERE clause:

| Operator | Description |
| :--- | ---: |
| = | Equal	
| > | Greater than	
| < | Less than	
| >= | Greater than or equal	
| <= | Less than or equal	
| <> | Not equal. Note: In some versions of SQL this operator may be written as !=	
| BETWEEN | Between a certain range	
| LIKE | Search for a pattern	
| IN | To specify multiple possible values for a column
