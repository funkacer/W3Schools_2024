```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_labels(line):
    x,y=line.get_data()
    labels=map(':'.join,zip(map(lambda s: '%g'%s,x),map(lambda s: '%g'%s,y)))
    list(map(plt.text,x,y,labels))
```

```python
x = np.arange(0, 10)
y = np.exp2(x)
line,= plt.plot(x, y, marker = "x")
#print(line.get_data())
#x,y=line.get_data()
#labels=map(','.join,zip(map(lambda s: '%g'%s,x),map(lambda s: '%g'%s,y)))
#print(list(map(plt.text,x,y,labels)))
add_labels(line)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```
