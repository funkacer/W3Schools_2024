```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = df.loc[x, "Duration"] / 10 #45 misto 450
```

```python
df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')

plt.show()
```
