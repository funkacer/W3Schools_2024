```python
import psycopg2
```

```python
# Connect to your postgres DB
conn = psycopg2.connect("dbname=test user=funka password=funka")
```

```python
# Open a cursor to perform database operations
cur = conn.cursor()
```

AS, ||
 - SQL aliases are used to give a table, or a column in a table, a temporary name.

Using aliases for columns:

```python
# Execute a query
cur.execute('''SELECT customer_id AS id
FROM customers;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```

AS is Optional

Same result without AS:

```python
# Execute a query
cur.execute('''SELECT customer_id id
FROM customers;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```

Concatenate two fields and call them product:

```python
# Execute a query
cur.execute('''SELECT product_name || unit AS product
FROM products;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```

Note: In the result of the example above we are missing a space between product_name and unit. To add a space when concatenating, use || ' ' ||.

```python
# Execute a query
cur.execute('''SELECT product_name || ' ' || unit AS product
FROM products;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```

If you want your alias to contain one or more spaces, like "My Great Products", surround your alias with double quotes.

```python
# Execute a query
cur.execute('''SELECT product_name AS "My Great Products"
FROM products;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```

```python
# Execute a query
cur.execute('''SELECT CONCAT (product_name, ' ', unit) AS "My Great Products"
FROM products;''')
```

```python
# Retrieve query results
records = cur.fetchall()
```

```python
records
```
