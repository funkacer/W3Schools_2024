```python
# visualize logs
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.linspace(0.01,10, 100)
```

```python
plt.plot(xpoints, np.log2(xpoints), linestyle = 'dotted', color = 'orange')
plt.plot(xpoints, np.log(xpoints), linestyle = 'dotted', color = 'blue')
plt.plot(xpoints, np.log10(xpoints), linestyle = 'dotted', color = 'green')
plt.grid(visible = True)
plt.show()
```
